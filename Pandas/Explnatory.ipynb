```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
df_best['budget_musd'] = df_best['budget_musd'].fillna(0)
df_best['revenue_musd'] = df_best['revenue_musd'].fillna(0)
```

```python
df.info()
```

```python
def best_worst(n,by,ascending=False ,min_bud = 0 , min_vote=0):
    df2 = df_best.loc[(df_best.budget_musd >= min_bud) &(df_best.vote_count > min_vote),
                      ["",by]].sort_values(by = by , ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

__Movies Top 5 - Highest Revenue__

```python
Highest_Revenue = df_best.sort_values(by='revenue_musd' , ascending=False).iloc[:5 ,:5]
HTML(Highest_Revenue.to_html(escape=False))
```

```python
Highest_Budget = df_best.sort_values(by='budget_musd' , ascending=False).iloc[:5 , :5]
HTML(Highest_Budget.to_html(escape=False))
```

```python
df_best['Profiet'] = df_best['revenue_musd'] - df_best['budget_musd']
highest_profiet_df = df_best.sort_values(by='Profiet', ascending=False).head(5)
HTML(highest_profiet_df.to_html(escape=False))
```

```python
mask_genres = df.genres.str.contains('Action') & df.genres.str.contains('Science Fiction')
mask_actor = df.cast.str.contains('Bruce Willis')
filtered = df[mask_actor & mask_genres][['title', 'vote_average']]
sorted_filtered = filtered.sort_values(by='vote_average', ascending=False)
sorted_filtered
```

```python
mask_director = df.director == 'Quentin Tarantino'
mask_actor2 = df.cast.str.contains('Uma Thurman')
quentin = df.loc[mask_director & mask_actor2, ['title', 'poster_path', 'runtime']]
HTML(quentin.sort_values(by='runtime').set_index('title').to_html(escape=False))

```

```python
mask_studio = df.production_companies.str.contains('Pixar') & df.production_companies.notna()
mask_time = df.release_date.between('2010-01-01' , '2015-12-31')
pixar = df.loc[mask_studio & mask_time,['title','poster_path','revenue_musd','release_date']].sort_values(by='revenue_musd' , ascending=False).head()
HTML(pixar.to_html(escape=False))
```

```python
mask_category = df.genres.str.contains('Action') | df.genres.str.contains('Thriller')
mask_language = df.original_language == 'en'
mask_vote = df.vote_average >= 7.5
mask_vote2 = df.vote_count >= 10
movie_suggestion = df.loc[mask_category&mask_language&mask_vote&mask_vote2&mask_vote2 ,
                          ['title' ,'poster_path','genres','vote_average']].sort_values(by='vote_average' , ascending=False).set_index('title').head()
HTML(movie_suggestion.to_html(escape=False))
```
